```python
def top_10(dataframe, column):
    """
    Returns the top 10 most frequent values in a specified column of a DataFrame.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing the data.
    - column (str): The name of the column in which to find the top values.

    Returns:
    - Series: A Series containing the top 10 most frequent values in the specified column,
      along with their respective counts, sorted in descending order of frequency.

    Example:
    top_10_values = top_10(df, 'column_name')
    """
    return dataframe[column].value_counts().sort_values(ascending=False).head(10)


def bottom_10(dataframe, column):
    """
    Returns the bottom 10 most frequent values in a specified column of a DataFrame.

    Parameters:
    - dataframe (DataFrame): The DataFrame containing the data.
    - column (str): The name of the column in which to find the top values.

    Returns:
    - Series: A Series containing the bottom 10 most frequent values in the specified column,
      along with their respective counts, sorted in descending order of frequency.

    Example:
    top_10_values = top_10(df, 'column_name')
    """
    return dataframe[column].value_counts().sort_values(ascending=False).tail(10)
```

```python
# Get the bottom 10 most frequent values in the 'categoryName' column of the DataFrame
bottom_10(ac, 'categoryName')
```

```python
def is_duplicated(dataframe):
    """
    Checks if there are any duplicate rows in the given DataFrame.

    Parameters:
    - dataframe (DataFrame): The DataFrame to check for duplicates.

    Returns:
    - tuple or str: If duplicate rows are found, returns a tuple containing the string
      'Duplicate rows found.' and a DataFrame containing the duplicated rows. If no
      duplicate rows are found, returns the string 'No duplicate rows found.'

    Example:
    is_duplicated_result = is_duplicated(df)
    if isinstance(is_duplicated_result, tuple):
        print(is_duplicated_result[0])
        print(is_duplicated_result[1])
    else:
        print(is_duplicated_result)
    """
    duplicated = dataframe[dataframe.duplicated()]
    if duplicated.empty:
        return 'No duplicate rows found.'
    else:
        return 'Duplicate rows found.', duplicated
```

```python
# checking for duplicate rows
is_duplicated(ac)
```

```python
# Initialize empty lists to store numeric and object column names
numeric_columns = []
object_columns = []

# Iterate over columns in the DataFrame
for column in ac:
    # Check if the column's data type is object
    if ac[column].dtype == 'object':
        # If it's an object column, add it to the object_columns list
        object_columns.append(column)
    else:
        # If it's not an object column, add it to the numeric_columns list
        numeric_columns.append(column)

# Create separate DataFrames for numeric and object columns
num_df = ac[numeric_columns]
obj_df = ac[object_columns]
```

```python
# Compute the correlation matrix for the numeric columns in the DataFrame
num_corr = num_df.corr()
```

```python
# Import colormaps from matplotlib
from matplotlib import colormaps

# Create a heatmap of the correlation matrix 'num_corr' using Seaborn
sns.heatmap(num_corr, annot=True, cmap='magma', fmt=".2f", annot_kws={"size": 10})
```
